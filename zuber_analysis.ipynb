```python
import pandas as pd

df_07 = pd.read_csv("moved_project_sql_result_07.csv")

df_07.head()
```

```python

df_07['start_ts'] = pd.to_datetime(df_07['start_ts'])

df_07['day_of_week'] = df_07['start_ts'].dt.dayofweek

df_07[['start_ts', 'day_of_week']].head()
```

```python

saturdays = df_07[df_07['day_of_week'] == 5]

rainy_saturdays = saturdays[saturdays['weather_conditions'] == 'Bad']
non_rainy_saturdays = saturdays[saturdays['weather_conditions'] == 'Good']

print("Viajes en sábados lluviosos:", len(rainy_saturdays))
print("Viajes en sábados sin lluvia:", len(non_rainy_saturdays))
```

```python
from scipy import stats

rainy_durations = rainy_saturdays['duration_seconds']
non_rainy_durations = non_rainy_saturdays['duration_seconds']

levene_test = stats.levene(rainy_durations, non_rainy_durations)
print("Prueba de Levene (p-valor):", levene_test.pvalue)

equal_var = levene_test.pvalue > 0.05

t_test = stats.ttest_ind(rainy_durations, non_rainy_durations, equal_var=equal_var)
print("Prueba t de Student (p-valor):", t_test.pvalue)
```

<b>Comentario del revisor</b> <a class="tocSkip"></a>

Gran trabajo con el desarrollo de la prueba de hipótesis. Además, lo complementas con la prueba de levene para mostrar el supuesto de varianzas iguales.

### Conclusión de la hipótesis

- Se hizo una prueba de Levene para verificar la igualdad de varianzas entre los viajes de sábados lluviosos y no lluviosos.
- El p-valor fue menor a 0.05, por lo que se asumió que las varianzas **no son iguales**.
- Luego se hizo una prueba t de Student con `equal_var=False`. También dio un p-valor < 0.05.

 **Conclusión:** hay evidencia estadísticamente significativa para rechazar la hipótesis nula.  
Por lo tanto, la duración promedio de los viajes **sí cambia** los sábados lluviosos.
